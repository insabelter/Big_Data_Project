```python
# Import some necessary modules

from kafka import KafkaConsumer
from pymongo import MongoClient
import json

# Connect to MongoDB and pizza_data database
client = MongoClient('localhost',27017)
db = client.cities.munich
print('connect consumer')
# connect kafka consumer to desired kafka topic
consumer = KafkaConsumer('testCSV',bootstrap_servers=['localhost:9092'])
print('connected consumer')
# Parse received data from Kafka
for msg in consumer:
    print(msg.value.rstrip())
    record = msg.value.decode().split(';')
    loc = record[0]
    tom = record[1]
    tom = tom[0:10]
    day = record[2]
    pedcount = record[3]
    temp = record[4]
    weather_cond = record[5]
    incidents = record[6]
    
    citylog = {'location':loc,
               'time of measurement':tom,
               'weekday':day,
               'pedestrians count':pedcount,
              # 'temperature in celsius':temp,
               'weather condition':weather_cond,
               'incidents':incidents
              }
    rec_id1 = db.insert_one(citylog)
    print("Data inserted with record ids", rec_id1)
```
